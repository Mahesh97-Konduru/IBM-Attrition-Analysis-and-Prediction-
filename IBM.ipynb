# Exploratory data analysis

```python
def diagnostic_plots(df, variable):
    
    plt.figure(figsize=(16,4))
    
    #histogram
    plt.subplot(1,3,1)
    sns.distplot(df[variable], bins=30)
    
     #q-q plot
    
    plt.subplot(1,3,2)
    stats.probplot(df[variable], dist ='norm', plot = plt)
    
    # box plot
    
    plt.subplot(1,3,3)
    sns.boxplot(y = df[variable])
    plt.show()
```

```python
for col in num_cols:
    diagnostic_plots(data,col)
    plt.clf()
    
```

# one hot encoding for categorical features

```python
data_ohehot = pd.get_dummies(data[cat_cols],drop_first=True)
```

```python
data_ohehot
```

```python
data2 = data.drop(cat_cols,axis=1)
```

```python
data_new = pd.concat((data_ohehot,data2),axis=1)
```

```python
data_new.head()
```

# Random Forest

# Without hyper parameter tuning

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score, recall_score, precision_score
from sklearn.model_selection import GridSearchCV
```

```python
X = data_new.drop(columns= ['Attrition'],axis=1)
y = data_new['Attrition']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state = 10)
```

```python
X_train.shape,y_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=12)
```

```python
# fit the model
model.fit(X_train, y_train)

# Get predictions
y_true, y_pred = y_test, model.predict(X_test)

# Evaluate train and test scores
train_ = model.score(X_train, y_train)
test_ = model.score(X_test, y_test)
                      
```

```python
# Other scores
accuracy = accuracy_score(y_true, y_pred).round(3)
            
precision = precision_score(y_true, y_pred).round(3)
            
recall = recall_score(y_true, y_pred).round(3)
```

```python
print('-'*20)
print('Training Score: {0:.3f}'.format(train_))
print('Test Score: {0:.3f}'.format(test_))
            
print('Accuracy Score: {0:.3f}'.format(accuracy))
print('Precision Score : {0:.3f}'.format(precision))
print('Recall Score : {0:.3f}'.format(recall))
```

# with hyper parameter tuning

```python
# Random Forest Classifier with hyperparameter tuning
model_hyper = RandomForestClassifier()
params = {'n_estimators' : [10, 20, 30, 50, 75, 100], 'max_depth': [1, 2, 3, 5, 7, 10]}

model_grid = GridSearchCV(model_hyper, param_grid = params)

model_grid.fit(X, y)

print(model_grid.best_estimator_)
print(model_grid.best_params_)
```

# Oversampling

```python
# Oversampling
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE(sampling_strategy = 'minority')
X_train_res, y_train_res = sm.fit_sample(X_train, y_train)
```

```python
X_train_res.shape,y_train_res.shape
```

```python
model = RandomForestClassifier(n_estimators=12)
```

```python
# fit the model
model.fit(X_train_res, y_train_res)

# Get predictions
y_true, y_pred = y_test, model.predict(X_test)

# Evaluate train and test scores
train_ = model.score(X_train_res, y_train_res)
test_ = model.score(X_test, y_test)
                      
```
